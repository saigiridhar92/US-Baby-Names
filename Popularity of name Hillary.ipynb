```python
# Plot Hillary
import matplotlib.pyplot as plt
from pylab import rcParams
import collections
from collections import defaultdict
import operator
import csv
```

```python
% matplotlib inline
Allstatenames = csv.reader(open("C:/Users/tsaig/OneDrive/Documents/00 - Data Sets/US Baby Names/output/StateNames.csv"))
```

```python
Names = ["Hillary"]
specificName = {}
```

```python
for x in Allstatenames:
    state = x[4]
    gender = x[3]
    year = x[2]
    name = x[1]
    count = x[5]
    
    if (name in Names):
        specificName[year] = specificName.get(year,0) + int(count)
```

```python
# Charts
specificName = collections.OrderedDict(sorted(specificName.items()))
```

```python
rcParams['figure.figsize'] = 18,8
plt.bar(range(len(specificName)), specificName.values())
plt.xticks(range(len(specificName)), specificName.keys())
plt.xticks(rotation = 90)
plt.title("Popularity of Name Hillary")
plt.show()
```
